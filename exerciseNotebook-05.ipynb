```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams

import scipy
from scipy.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8, 4
sns.set_style("whitegrid")
```

```python
sns.pairplot(cars)
```

### Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coefficient, p_value = pearsonr(mpg, hp)
print('PearsonR Correlation Coefficient %0.3f' % (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, wt)
print('PearsonR Correlation Coefficient %0.3f' % (pearsonr_coefficient))
```

## 05_05

## Delving into non-parametric methods using pandas and scipy

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams

import scipy
from scipy.stats import spearmanr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 14, 7
sns.set_style("whitegrid")
```

### The Spearman Rank Correlation

```python
#address = '/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/mtcars.csv'

address = './data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
```

```python
sns.pairplot(cars)
```

```python
x = cars[['cyl', 'vs', 'am', 'gear']]
sns.pairplot(x)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl, vs)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, am)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, gear)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

### Chi-square test for independence

```python
table = pd.crosstab(cyl, am)
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic % 0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, vs)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic % 0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, gear)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic % 0.3f p_value %0.3f' % (chi2, p))
```

## Extreme value analysis using univariate methods

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
#address = '/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/iris.data.csv'
address = './data/iris.data.csv'

df= pd.read_csv(filepath_or_buffer=address, header=None, sep=',')

df.columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Species']
x = df.iloc[:,0:4].values
y = df.iloc[:,4].values
df[:5]
```

### Identifying outliers from Tukey boxplots

```python
df.boxplot(return_type='dict')
plt.plot()
```

```python
Sepal_width = x[:,1]
iris_outliers = (Sepal_width>4)
df[iris_outliers]
```
